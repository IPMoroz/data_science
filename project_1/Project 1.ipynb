2. Теперь нас интересует столбец **"Пол, возраст"**. Сейчас он представлен в формате **<Пол , возраст , дата рождения >**. Например:
* Мужчина , 39 лет , родился 27 ноября 1979 
* Женщина , 21 год , родилась 13 января 2000
Как вы понимаете, нам необходимо выделить каждый параметр в отдельный столбец.

Создайте два новых признака **"Пол"** и **"Возраст"**. При этом важно учесть:
* Признак пола должен иметь 2 уникальных строковых значения: 'М' - мужчина, 'Ж' - женщина. 
* Признак возраста должен быть представлен целыми числами.

Выполните преобразование, ответьте на контрольные вопросы и удалите признак **"Пол, возраст"** из таблицы.

*Совет: обратите внимание на структуру текста в столбце, в части на то, как разделены параметры пола, возраста и даты рождения между собой - символом ' , '. 
Гарантируется, что структура одинакова для всех строк в таблице. Вы можете воспользоваться этим.*


```python
def get_gender(social_info):
    
    social_list = social_info.split(',')
    #задаю начальное значение пола (7 знаков от начала строки)
    gender = social_list[0][:7]
    #задаю условие выбор буквы
    if gender == 'Мужчина':
        gender = 'М'
    else:
        gender = 'Ж'
    return gender

def get_age(social_info):
    
    social_list = social_info.split(',')
    #выбираю второе значение в списке (информация о возрасте)
    age_str = social_list[1]
    #разбиваю строку по разделителю
    age_list = age_str.split(' ')
    #выбираю первые два знака обозначающие числовое значение возраста и привожу к целым числам
    age = int(age_list[2])
    
    return age

#применяю функции при создании новых столбцов
hh_df['Пол'] = hh_df['Пол, возраст'].apply(get_gender)
hh_df['Возраст'] = hh_df['Пол, возраст'].apply(get_age)
#удаляю лишнее
hh_df.drop(['Пол, возраст'], axis=1, inplace=True)
```

```python
mask = hh_df['Пол'] == 'Ж'
female_resumes = round(((hh_df[mask]['Пол'].count() / hh_df['Пол'].count())*100), 2)
middle_ages = round((hh_df['Возраст'].mean()), 1)
print(f'В наших данных представлено {female_resumes} % женских резюме')
print(f'Средний возраст соискателей равен {middle_ages} лет')
```

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:

* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
def get_work_experience(work):
    #задаю списки слов для дальнейшего распределения
    year_sign = ['год', 'года', 'лет', ]
    month_sign = ['месяц', 'месяца', 'месяцев']
    #создаю пустой список
    work_experience = []
    
    #первое условие на принадлежность значения NaN
    if work is np.NaN:
        work_experience = work
    else:
        #привожу значение к типу строка, для применения метода split
        work = str(work)
        work_list = work.split(' ')
        #второе условие по первому значению в списке (меня интересуют занчения формата "Опыт работы ...", остальное приводим к NaN)
        if work_list[0] != 'Опыт':
            work_experience = np.NaN
        else:
            #третье условие по вхождению в ранее заданные списки, приводим значения к целым числам
            if work_list[3] in month_sign:
                work_experience = int(work_list[2])
            elif work_list[3] in year_sign and work_list[5] in month_sign:
                work_experience = (int(work_list[2]) * 12) + int(work_list[4])
            else:
                work_experience = int(work_list[2]) * 12
    return work_experience


hh_df['Опыт работы (месяц)'] = hh_df['Опыт работы'].apply(get_work_experience)
#удаляю лишнее
hh_df.drop(['Опыт работы'], axis=1, inplace=True)
```

```python
work_experience = hh_df['Опыт работы (месяц)'].median()
print(f'Медианный опыт работы (в месяцах) равен {work_experience} месяцев')
```

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
def get_city(readiness):
    
    #задаю список столичных городов
    capital_cities = [
        'Москва',
        'Санкт-Петербург'
    ]
    #задаю список городов миллионников
    million_cities = [
        'Новосибирск',
        'Екатеринбург',
        'Нижний Новгород',
        'Казань',
        'Челябинск',
        'Омск',
        'Самара',
        'Ростов-на-Дону',
        'Уфа',
        'Красноярск',
        'Пермь',
        'Воронеж',
        'Волгоград'
    ]
    ready_list = readiness.split(',')
    #на первой позиции находится обозначение города, делим строку по пробелу (причина: наличие конструкции "город (область)")
    city_list = ready_list[0].split(' ')
    city = []
    
    #проверяем условие соответствия списку столичных городов
    if city_list[0] in capital_cities:
        city = city_list[0]
    #для соответствия списку миллионников задаем особые условия для "Нижнего Новгорода" по первому и второму слову
    elif city_list[0] == 'Нижний':
        if city_list[0]+' '+city_list[1] in million_cities:
            city = 'Нижний Новгород'
        else:
            city = 'другие'
    #проверяем условие соответствия списку миллионников
    elif city_list[0] in million_cities:
        city = city_list[0]
    else:
        city = 'другие'
    
    return city

def get_relocate(readiness):
    
    #задаю  список ключевых слов
    check_list = ['готов', 'готова', 'хочу']
    #перед разделением строки по "," избавляюсь от лишних пробелов, которые будут мешать в дальнейшем
    ready_list = readiness.replace(', ', ',').split(',')
    relocate_list = []
    relocate = []
    
    #прохожу по разделенной строке для создания списка списков
    for string in ready_list:
        relocate_list.append(string.split(' '))

    #признак готовности расположен на 2, либо 3 месте. Поэтому прохожу по очереди слева-направо, пока не найду ключевые слова, либо слово "не"
    for phrase in relocate_list:
        for word in phrase:
            if word == 'не':
                return False
            elif word in check_list:
                return True


def get_bussiness_trip(readiness):
    
    ready_list = readiness.replace(', ', ',').split(',')
    #признак готовности первый справа
    bt_list = ready_list[-1].split(' ')
    bt = []
    
    if bt_list[0] == 'не':
        bt = False
    else:
        bt = True
    return bt


hh_df['Город'] = hh_df['Город, переезд, командировки'].apply(get_city)
hh_df['Готовность к переезду'] = hh_df['Город, переезд, командировки'].apply(get_relocate)
hh_df['Готовность к командировкам'] = hh_df['Город, переезд, командировки'].apply(get_bussiness_trip)
#удаляю лишнее
hh_df.drop(['Город, переезд, командировки'], axis=1, inplace=True)
```

```python
proportion = round((hh_df[hh_df['Город']=='Санкт-Петербург'].shape[0] / hh_df.shape[0])*100)
print(f'В Санкт-Петербурге живет {proportion} % соискателей')
mask1 = hh_df['Готовность к переезду']==True
mask2 = hh_df['Готовность к командировкам']==True
readiness = round((hh_df[mask1&mask2].shape[0]/hh_df.shape[0])*100)
print(f'Одновременно и к переездам, и к командировкам готовы {readiness} % соискателей')
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
def get_employment(set_categories):
    #задаю базовый список при прохождении по элементам которого определяется соответствие исследуемого списка
    emp_base = ['полная занятость', 'частичная занятость', 'проектная работа', 'волонтерство', 'стажировка']
    #исследуемый список, без лишних пробелов
    emp_category = set_categories.replace(', ', ',').split(',')
    employment = []
    
    #прохожу по элементам базового списка и возвращаю True при совпадении, и False при отсутствии
    for phrase in emp_base:
        if phrase in emp_category:
            employment.append(True)
        else:
            employment.append(False)
    return employment


def get_schedule(set_categories):
    
    sch_base = ['полный день', 'сменный график', 'гибкий график', 'удаленная работа', 'вахтовый метод']
    
    sch_category = set_categories.replace(', ', ',').split(',')
    schedule = []
    
    for phrase in sch_base:
        if phrase in sch_category:
            schedule.append(True)
        else:
            schedule.append(False)
    return schedule

#применяю функции к исходным столбцам. Из создаваемого базового объекта Series при помощи apply извлекаю значение соответствующее этому признаку 
hh_df['полная занятость'] = hh_df['Занятость'].apply(get_employment).apply(lambda x: x[0])
hh_df['частичная занятость'] = hh_df['Занятость'].apply(get_employment).apply(lambda x: x[1])
hh_df['проектная работа'] = hh_df['Занятость'].apply(get_employment).apply(lambda x: x[2])
hh_df['волонтерство'] = hh_df['Занятость'].apply(get_employment).apply(lambda x: x[3])
hh_df['стажировка'] = hh_df['Занятость'].apply(get_employment).apply(lambda x: x[4])
hh_df['полный день'] = hh_df['График'].apply(get_schedule).apply(lambda x: x[0])
hh_df['сменный график'] = hh_df['График'].apply(get_schedule).apply(lambda x: x[1])
hh_df['гибкий график'] = hh_df['График'].apply(get_schedule).apply(lambda x: x[2])
hh_df['удаленная работа'] = hh_df['График'].apply(get_schedule).apply(lambda x: x[3])
hh_df['вахтовый метод'] = hh_df['График'].apply(get_schedule).apply(lambda x: x[4])

hh_df.drop(['Занятость', 'График'], axis=1, inplace=True)
```

```python
mask1 = hh_df['проектная работа'] == True
mask2 = hh_df['волонтерство'] == True
mask3 = hh_df['вахтовый метод'] == True
mask4 = hh_df['гибкий график'] == True
sign1 = hh_df[mask1&mask2].shape[0]
sign2 = hh_df[mask3&mask4].shape[0]
print(f'Проектную работу и волонтёрство ищут {sign1} людей')
print(f'Работать вахтовым методом и с гибким графиком хотят {sign2} людей')
```

В чем наша беда? В том, что помимо желаемой заработной платы соискатель указывает валюту, в которой он бы хотел ее получать, например:

Создайте новый DataFrame из полученного файла. В полученной таблице нас будут интересовать столбцы:
* "currency" - наименование валюты в ISO кодировке,
* "date" - дата, 
* "proportion" - пропорция, 
* "close" - цена закрытия (последний зафиксированный курс валюты на указанный день).


Перед вами таблица соответствия наименований иностранных валют в наших данных и их общепринятых сокращений, которые представлены в нашем файле с курсами валют. Например, для казахстанского тенге курс на 20.08.2019 составляет 17.197 руб. за 100 тенге, тогда итоговый курс равен - 17.197 / 100 = 0.17197 руб за 1 тенге.

Осталось только понять, откуда брать дату, по которой определяется курс? А вот же она - в признаке **"Обновление резюме"**, в нем содержится дата и время, когда соискатель выложил текущий вариант своего резюме. Нас интересует только дата, по ней бы и будем сопоставлять курсы валют.

Теперь у нас есть вся необходимая информация для того, чтобы создать признак "ЗП (руб)" - заработная плата в рублях.

После ответа на контрольные вопросы удалите исходный столбец заработной платы "ЗП" и все промежуточные столбцы, если вы их создавали.

Итак, давайте обсудим возможный алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
#создаю DataFrame из скачанного файла, привожу столбец date к формату datetime
#достаю дату при помощи признака, удаляю лишние столбцы
ExchangeRates = pd.read_csv('data/ExchangeRates.csv')
ExchangeRates['date'] = pd.to_datetime(ExchangeRates['date'])
ExchangeRates['date'] = ExchangeRates['date'].dt.date
ExchangeRates.drop(['per', 'time', 'vol'], axis=1, inplace=True)
#привожу столбец date к формату datetime, достаю дату при помощи признака
hh_df['Обновление резюме'] = pd.to_datetime(hh_df['Обновление резюме'], dayfirst=True)
hh_df['Дата'] = hh_df['Обновление резюме'].dt.date

def get_salary(desired):
    
    salary_list = desired.split(' ')
    
    return float(salary_list[0])


def get_currency(desired):
    
    currency_dict = {
        'грн':'UAH',
        'USD':'USD',
        'EUR':'EUR',
        'белруб':'BYN',
        'KGS':'KGS',
        'сум':'UZS',
        'AZN':'AZN',
        'KZT':'KZT',
        'руб':'RUB'
    }
    currency_list = desired.replace('.', '').split(' ')
    currency = []
    
    if currency_list[-1] in currency_dict.keys():
        currency = currency_dict[currency_list[-1]]
    
    return currency
    

hh_df['Salary'] = hh_df['ЗП'].apply(get_salary)
hh_df['Currency'] = hh_df['ЗП'].apply(get_currency)
#для удобства слияния таблиц, переименую столбцы в резюме
ExchangeRates.rename(columns={'currency':'Currency', 'date':'Дата'}, inplace=True)
#объединяю таблицы методом merge (тип left) по двум признакам
hh_df = hh_df.merge(
    ExchangeRates,
    on=['Currency', 'Дата'],
    how='left'
)
#заполняю пустые значения единицами
hh_df = hh_df.fillna({'close':1, 'proportion':1})
#создаю новый столбец
hh_df['ЗП (руб)'] = hh_df['Salary'] * hh_df['close'] / hh_df['proportion']

hh_df.drop(['ЗП', 'Обновление резюме', 'Salary', 'Currency', 'close', 'proportion'], axis=1, inplace=True)
```

```python
desired_salary = hh_df['ЗП (руб)'].median()
print(f'Желаемая медианная заработная плата соискателей равна {desired_salary} рублей')
```

Выведите таблицу с полученными выбросами и оцените, с каким возрастом соискатели попадают под категорию выбросов?

```python
log_age = np.log(hh_df['Возраст'])

fig, ax = plt.subplots(1, 1, figsize=(8, 4))

histplot = sns.histplot(log_age, bins=30, ax=ax)
histplot.axvline(log_age.mean(), color='k', lw=2)
histplot.axvline(log_age.mean()+ 3 * log_age.std(), color='k', ls='--', lw=2)
histplot.axvline(log_age.mean()- 3 * log_age.std(), color='k', ls='--', lw=2)
histplot.set_title('Логарифмическое распределение признака возраст');
```

```python
def outliers_z_score_mod(data, feature, log_scale=False, left=3, right=3):
    if log_scale:
        x = np.log(data[feature])
    else:
        x = data[feature]
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - left * sigma
    upper_bound = mu + right * sigma
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x > lower_bound) & (x < upper_bound)]
    return outliers, cleaned

outliers, cleaned = outliers_z_score_mod(hh_df, 'Возраст', log_scale=True, right=4)
print(f'Число выбросов по методу z-отклонения: {outliers.shape[0]}')
print(f'Результирующее число записей: {cleaned.shape[0]}')
outliers
```

```python
#удаляем выбросы
hh_df = hh_df.drop(outliers.index, axis=0)
```

Логарифмическое распределение имеет правостороннюю ассимметрию (справа значение больше). При послаблении справа на 1 сигму получаем 3 выброса - это два 15-летних соискателя и один старичек 100 летний.
